```python
import pandas as pd

# Load project data
task_df = pd.read_csv("Downloads/OneDrive_2025-03-11/extracted csv/TASK.csv", encoding="latin1")
task_pred_df = pd.read_csv("Downloads/OneDrive_2025-03-11/extracted csv/TASKPRED.csv", encoding="latin1")
task_rsrc_df = pd.read_csv("Downloads/OneDrive_2025-03-11/extracted csv/TASKRSRC.csv", encoding="latin1")
task_cost_df = pd.read_csv("Downloads/OneDrive_2025-03-11/extracted csv/PROJCOST.csv", encoding="latin1")
resource_capacity_df = pd.read_excel("Downloads/OneDrive_1_11-03-2025/Resource Capacity.xlsx")

# Initialize the task graph
task_graph = {}

# Populate the task graph with data from the CSV files
for index, row in task_rsrc_df.iterrows():
    project_id = row['proj_id']
    task_id = row['task_id']
    act_start_date = pd.to_datetime(row['act_start_date'])
    act_end_date = pd.to_datetime(row['act_end_date'])
    today = pd.to_datetime("today")

    # Calculate completion percentage based on actual start and end dates
    if pd.isna(act_start_date) or pd.isna(act_end_date):
        completion = 0
    elif today >= act_end_date:
        completion = 100
    elif today <= act_start_date:
        completion = 0
    else:
        total_duration = (act_end_date - act_start_date).days
        elapsed_duration = (today - act_start_date).days
        completion = (elapsed_duration / total_duration) * 100

    if project_id not in task_graph:
        task_graph[project_id] = {'tasks': []}

    task_graph[project_id]['tasks'].append({'task_id': task_id, 'completion': completion})

# Define the function to calculate percentage completion
def calculate_percentage_completion(task_graph):
    project_completion = {}
    for project_id, project_data in task_graph.items():
        total_completion = sum(task['completion'] for task in project_data['tasks'])
        num_tasks = len(project_data['tasks'])
        project_completion[project_id] = {
            'completion_percentage': total_completion / num_tasks if num_tasks > 0 else 0
        }
    return project_completion

# Compute and display percentage completion per project
task_completion_status = calculate_percentage_completion(task_graph)
for project, data in task_completion_status.items():
    print(f"Project: {project} - {data['completion_percentage']:.2f}% Completed")
```
